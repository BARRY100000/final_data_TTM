```python
import pandas as pd

# 讀取預測結果和實際結果文件
predictions = pd.read_csv('content/result/content/nba_predictions.csv')
actuals = pd.read_csv('content/result/content/final_data_sec_nonzero_sufficient.csv')

# 合併資料，使用 'playerid' 和 'gameid' 作為鍵
merged_data = pd.merge(predictions, actuals, on=['playerid', 'gameid'], suffixes=('_pred', '_actual'))

# 定義不參與分析的欄位
label_columns = ['date', 'playerid', 'gameid', 'home_basic', 'home', 'win', 'team', 'name', 'date_num']

# 找出共有的數值欄位，排除 label_columns 中的欄位
common_columns = set(predictions.columns).intersection(set(actuals.columns)) - set(label_columns)

# 建立一個空的 DataFrame 來儲存分析結果
error_analysis_results = pd.DataFrame(columns=['column', 'average_error', 'std_error'])

# 計算每個共有欄位的誤差，並記錄到 error_analysis_results 中
for column in common_columns:
    # 確保欄位資料型別為數值
    merged_data[f"{column}_pred"] = pd.to_numeric(merged_data[f"{column}_pred"], errors='coerce')
    merged_data[f"{column}_actual"] = pd.to_numeric(merged_data[f"{column}_actual"], errors='coerce')
    merged_data[f"{column}_error"] = merged_data[f"{column}_pred"] - merged_data[f"{column}_actual"]
    
    # 計算平均誤差和標準差
    average_error = merged_data[f"{column}_error"].mean()
    std_error = merged_data[f"{column}_error"].std()
    
    # 將結果添加到 DataFrame
    new_row = pd.DataFrame({'column': [column], 'average_error': [average_error], 'std_error': [std_error]})
    error_analysis_results = pd.concat([error_analysis_results, new_row], ignore_index=True)

# 將分析結果儲存為 CSV 或 Excel 檔案
error_analysis_results.to_csv('error_analysis_results.csv', index=False)
error_analysis_results.to_excel('error_analysis_results.xlsx', index=False)

print("誤差分析結果已儲存至 'error_analysis_results.csv' 和 'error_analysis_results.xlsx'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取正確資料集和預測資料集
correct_data = pd.read_csv('final_data_sec_nonzero_sufficient.csv')
prediction_data = pd.read_csv('nba_predictions5.csv')

def plot_prediction_accuracy(player_id, column_name):
    """
    此函數會繪製出指定球員(player_id)在指定欄位(column_name)的趨勢折線圖。
    圖表內容包含正確資料集的10個連續值及其後的1個預測值，以此類推。
    
    參數:
        player_id: 球員的ID，將從資料集中過濾出對應的數據。
        column_name: 欲分析的數據欄位名稱。
    """
    # 過濾出指定球員的資料
    correct_data_player = correct_data[correct_data['playerid'] == player_id]
    prediction_data_player = prediction_data[prediction_data['playerid'] == player_id]
    
    # 檢查資料是否包含所選欄位
    if column_name not in correct_data.columns or column_name not in prediction_data.columns:
        print(f"選定的欄位 {column_name} 不存在於資料集中。")
        return

    # 初始化圖形
    plt.figure(figsize=(12, 6))
    
    # 設定移動窗口範圍：從正確資料集的第1行到倒數第10行
    for i in range(len(correct_data_player) - 9):
        # 抓取正確資料集中的10個連續值
        correct_values = correct_data_player[column_name].iloc[i:i+10].values
        # 抓取預測資料集中第(i+10)行的值，作為第11個數
        predicted_value = prediction_data_player[column_name].iloc[i + 10]
        
        # 合併正確數據和預測值
        values = list(correct_values) + [predicted_value]
        
        # 繪製折線圖
        plt.plot(range(i + 1, i + 12), values, marker='o', label=f'Window {i+1}' if i == 0 else "")
    
    # 設定圖表標題和標籤
    plt.title(f"Player ID {player_id} - {column_name} 正確與預測趨勢圖")
    plt.xlabel("Data Point Index")
    plt.ylabel(column_name)
    plt.legend(["正確資料集 (10值) + 預測值 (1值)"], loc="upper left")
    plt.grid(True)
    plt.show()

# 執行範例 - 替換 player_id 和 column_name 參數以進行分析
# 範例: plot_prediction_accuracy(player_id=1, column_name='指定的欄位名稱')
plot_prediction_accuracy(player_id=1, column_name='your_column_name')
```
