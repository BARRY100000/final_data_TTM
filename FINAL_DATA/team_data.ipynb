```python
import pandas as pd
data = pd.read_csv('team_data/merged_team_FINAL.csv')
# 假設 data_sorted 是你已經排序好的資料
data_sorted = data.copy()

# 獲取所有數值型特徵列，排除 'playerid' 和 'gameid'
columns_to_exclude = [ 'gameid', 'home', 'date', 'team']
features_to_replace = data_sorted.select_dtypes(include=['float64', 'int64']).columns
features_to_replace = [col for col in features_to_replace if col not in columns_to_exclude]

# 定義前幾場比賽的數據平均值，這個變量可以調整
num_games_to_average = 5  # 可以調整這個值，表示取前幾場的數據平均

# 從最後一行開始，往前尋找對應的 playerid 的前幾場比賽
for i in range(len(data_sorted) - 1, -1, -1):
    current_player = data_sorted.loc[i, 'team']

    # 保存找到的前幾場比賽數據的列表
    previous_games = []

    # 遍歷查找前幾場比賽
    for j in range(i - 1, -1, -1):
        if data_sorted.loc[j, 'team'] == current_player and data_sorted.loc[j, 'SEC'] > 0:
            previous_games.append(data_sorted.loc[j, features_to_replace])
        if len(previous_games) >= num_games_to_average:
            break

    # 如果找到的前幾場比賽數據不為空，計算平均值
    if len(previous_games) > 0:
        average_values = pd.DataFrame(previous_games).mean()
        data_sorted.loc[i, features_to_replace] = average_values
    else:
        # 如果沒有找到前幾場比賽，將當前行數據設為 0
        data_sorted.loc[i, features_to_replace] = 0

# 查看結果
data_sorted.head()
```

```python
grouped_data = pd.DataFrame(data_sorted)
grouped_data.to_csv('team_data/merged_team_FINAL_lookforward.csv', index=False)
```

```python
import pandas as pd
merged_df = data_sorted
# 假設已經合併的 DataFrame 是 merged_df
# 將 `home` 為 1 與 0 的資料分成兩個 DataFrame
home_true = merged_df[merged_df['home'] == 1].set_index('gameid')
home_false = merged_df[merged_df['home'] == 0].set_index('gameid')

# 使用 gameid 合併兩個 DataFrame，home 為 0 的資料會在右側
final_df = home_true.join(home_false, lsuffix='_home1', rsuffix='_home0')

# 儲存合併後的 DataFrame 到新的 CSV 檔案
output_path = "team_data/final_team_data/final_merged_team_lookforward.csv"
final_df.to_csv(output_path, index=True)

print(f"Final merged CSV saved to {output_path}")
```

```python
import pandas as pd

# 讀取 CSV 檔案
final_lookforward_df = pd.read_csv("team_data/final_team_data/final_merged_team_lookforward.csv")
final_data_sec_df = pd.read_csv("content/final_data_sec_nonzero_sufficient.csv")

# 根據 gameid 將 final_merged_team_lookforward 的資料合併到 final_data_sec_nonzero_sufficient 的右側
merged_final_data = final_data_sec_df.merge(final_lookforward_df, on='gameid', how='left')

# 儲存合併後的 DataFrame 至新的 CSV 檔案
output_path = "content/final_merged_with_lookforward.csv"
merged_final_data.to_csv(output_path, index=False)

print(f"合併完成並儲存為 {output_path}")
```
